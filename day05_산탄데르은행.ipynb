```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.2,
                                                    random_state=0)

train_cnt = y_train.count()
test_cnt = y_test.count()
train_cnt, test_cnt
```

## 학습 및 평가

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
```

```python
xgb_clf = XGBClassifier(n_estimators=500,random_state=156)

xgb_clf.fit(X_train,
            y_train,
            early_stopping_rounds = 100 ,
           eval_metric='auc',
           eval_set=[(X_train,y_train),(X_test,y_test)])

xgb_roc_score = roc_auc_score(y_test,xgb_clf.predict_proba(X_test)[:,1])
xgb_roc_score
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
xgb_clf = XGBClassifier(n_estimators=100)
params = {
'max_depth':[5,7],
'min_child_weight':[1,3],
'colsample_bytree': [0.5,0.75]}

gridcv = GridSearchCV(xgb_clf, param_grid=params, cv=3)
gridcv.fit(X_train,y_train,early_stopping_rounds=30,
          eval_metric='auc',
          eval_set=[(X_train,y_train),(X_test,y_test)])

print('최적파라미터', gridcv.best_params_)
xgb_roc_score = roc_auc_score(y_test,gridcv.predict_proba(X_test)[:,1])
```

```python
xgb_roc_score = roc_auc_score(y_test,gridcv.predict_proba(X_test)[:,1])
xgb_roc_score
```

최적파라미터 {'colsample_bytree': 0.5, 'max_depth': 5, 'min_child_weight': 3} -->> 정확도 : 0.8444548670726981

```python
xgb_clf = XGBClassifier(n_estimators=1000,
                        learning_rate = 0.02,
                        max_depth= 5,
                        min_child_weight=3,
                        colsample_bytree=0.5,
                        reg_alpha=0.03,
                        random_state=156)

xgb_clf.fit(X_train,
            y_train,
            early_stopping_rounds = 200 ,
           eval_metric='auc',
           eval_set=[(X_train,y_train),(X_test,y_test)])

xgb_roc_score = roc_auc_score(y_test,xgb_clf.predict_proba(X_test)[:,1])
xgb_roc_score
```

최적파라미터 {'colsample_bytree': 0.5, 'max_depth': 5, 'min_child_weight': 3} -->> 정확도 : 0.8444548670726981

estimator를 100개로 증가시킨후 적용  -->>  정확도 : 0.846275892081954


```python
import matplotlib.pyplot as plt
%matplotlib inline

from xgboost import plot_importance

fig, ax = plt.subplots(figsize=(10,8))
plot_importance(xgb_clf, ax=ax, max_num_features=20, height = 0.4)
```

# LightGBM

```python
from lightgbm import LGBMClassifier
```

```python
lgbm_clf = LGBMClassifier(n_estimators=500)

evals = [(X_test, y_test)]
lgbm_clf.fit(X_train,y_train, early_stopping_rounds=100, eval_metric='auc', eval_set=evals,
            verbose=True)

lgbm_roc_score = roc_auc_score(y_test,lgbm_clf.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC : {0:.4f}'.format(lgbm_roc_score))
```

ROC AUC : 0.8409

```python
lgbm_clf = LGBMClassifier(n_estimators=200)

params = {'max_depth':[128,160],
         'min_child_samples': [60,100],
         'subsample':[0.8,1],
         'num_leaves':[32,64]}
gridcv = GridSearchCV(lgbm_clf,param_grid=params, cv=3)
gridcv.fit(X_train,
           y_train,
           early_stopping_rounds=30,
          eval_metric='auc',
          eval_set=[(X_train,y_train),(X_test,y_test)])

print('최적 파라미터 : ',gridcv.best_params_ )
roc_auc_score(y_test,gridcv.best_estimator_.predict_proba(X_test)[:,1])
```
